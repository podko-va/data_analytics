Парные диаграммы (pairplot)
🔹 Отображает зависимости между всеми числовыми признаками сразу (точечные графики + гистограммы).

📌 Что даёт?

Видны линейные зависимости между признаками.
Можно обнаружить скопления данных и аномалии.

Логарифмическое масштабирование (для выявления скрытых зависимостей)
Иногда линейные связи плохо видны, но можно попробовать логарифмы:

📌 Что даёт?

Позволяет убрать влияние выбросов и увидеть скрытые закономерности.
Данные могут лучше ложиться на линейную модель.

```python
import numpy as np
df["log_median_income"] = df["median_income"].apply(lambda x: np.log(x + 1))
df["log_median_house_value"] = df["median_house_value"].apply(lambda x: np.log(x + 1))

sns.scatterplot(x=df["log_median_income"], y=df["log_median_house_value"])
plt.title("Логарифмическая зависимость дохода и цены жилья")
plt.show()
```

```python
df["age_category"] = pd.cut(df["housing_median_age"], bins=[0, 20, 40, 60, 80, 100], labels=["0-20", "21-40", "41-60", "61-80", "80+"])

plt.figure(figsize=(10, 6))
sns.boxplot(x="age_category", y="median_house_value", data=df)
plt.title("Как возраст домов влияет на их цену")
plt.show()
```

4. Двумерная гистограмма (hexbin plot)
🔹 Хорошо показывает плотность данных при больших объёмах.

📌 Что даёт?

Позволяет видеть, где наибольшая концентрация значений.
Убирает "шум" и выбросы, выделяя основные закономерности.




```python
plt.figure(figsize=(8, 6))
plt.hexbin(df["median_income"], df["median_house_value"], gridsize=30, cmap="coolwarm", mincnt=1)
plt.colorbar(label="Количество домов")
plt.xlabel("Медианный доход")
plt.ylabel("Цена дома")
plt.title("Плотность цен в зависимости от дохода")
plt.show()
```

7. Анализ с PCA (поиск главных компонент)
Если данных много, можно сократить их размерность и выявить главные скрытые закономерности с PCA (анализ главных компонент).

📌 Что даёт?

Позволяет понять, какие признаки наиболее влияют на структуру данных.
Упрощает анализ и ускоряет машинное обучение.

```python

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Стандартизация данных
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[["median_income", "total_rooms", "housing_median_age", "median_house_value"]])

# PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Визуализация
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.5)
plt.xlabel("Первая главная компонента")
plt.ylabel("Вторая главная компонента")
plt.title("PCA - снижение размерности")
plt.show()
```
